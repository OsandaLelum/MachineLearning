```python
import numpy as np
import matplotlib.pyplot as plt

# Generate example data (replace this with your actual process data)
np.random.seed(42)
data = np.random.normal(loc=10, scale=2, size=50)  # Example data with a mean of 10 and a standard deviation of 2

# Calculate the control limits
sample_size = 5  # Number of data points in each sample
n_samples = len(data) // sample_size
x_bar = np.mean(data.reshape((n_samples, sample_size)), axis=1)
r = np.ptp(data.reshape((n_samples, sample_size)), axis=1)

# Control chart constants for a sample size of 5 (you can adjust these based on your process)
A2 = 0.729
D3 = 0
D4 = 2.282

# Calculate UCL, LCL, USL, and LSL
UCL_xbar = x_bar.mean() + A2 * r.mean()
LCL_xbar = x_bar.mean() - A2 * r.mean()
USL = 12  # Example Upper Specification Limit (replace with your process limit)
LSL = 8   # Example Lower Specification Limit (replace with your process limit)

# Plot the control chart
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(x_bar, label='X-bar')
plt.axhline(UCL_xbar, color='r', linestyle='--', label='UCL (X-bar)')
plt.axhline(LCL_xbar, color='g', linestyle='--', label='LCL (X-bar)')
plt.axhline(USL, color='b', linestyle='--', label='USL')
plt.axhline(LSL, color='m', linestyle='--', label='LSL')
plt.title('X-bar Chart')
plt.xlabel('Sample')
plt.ylabel('Value')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(r, label='R')
plt.axhline(D4 * r.mean(), color='r', linestyle='--', label='D4 * R-bar')
plt.title('R Chart')
plt.xlabel('Sample')
plt.ylabel('Range')
plt.legend()

plt.tight_layout()
plt.show()
```
