```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate data for tire layer thickness (in millimeters)
np.random.seed(0)
num_samples = 100
layer_thickness = np.random.normal(5, 0.2, num_samples)  # Mean thickness of 5mm with some variation

# Define upper and lower specification limits for layer thickness
USL = 5.5  # Upper Specification Limit
LSL = 4.5  # Lower Specification Limit

# Create a time series for the data (time steps)
time_steps = np.arange(num_samples)

# Plot the layer thickness data
plt.figure(figsize=(10, 5))
plt.plot(time_steps, layer_thickness, marker='o', linestyle='-')
plt.axhline(USL, color='r', linestyle='--', label='Upper Spec Limit')
plt.axhline(LSL, color='g', linestyle='--', label='Lower Spec Limit')
plt.title('Tire Layer Thickness Over Time')
plt.xlabel('Time Steps')
plt.ylabel('Layer Thickness (mm)')
plt.legend()
plt.grid(True)

# Detect anomalies (thickness outside specification limits)
anomalies = np.where((layer_thickness > USL) | (layer_thickness < LSL))[0]

# Display detected anomalies
if len(anomalies) > 0:
    print("Anomalies detected at time steps:", anomalies)
else:
    print("No anomalies detected.")

plt.show()
```
